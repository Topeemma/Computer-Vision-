# COMPUTER VISION
**Computer vision is a field of artificial intelligence that gives computers the ability to see, understand, and interpret the visual world
The goal of computer vision is to enable computers to see the world in the same way that humans do. This is a challenging task, as the visual world is complex and constantly changing. However, recent advances in artificial intelligence and machine learning have made significant progress in the field of computer vision**

**what is an Image**
> a digital image is not a continuous image, like a painting or a photograph. Instead, it is made up of a grid of pixels, and each pixel has a specific value that represents its color or brightness. It is an array of pixels arranged in columns and rows. Pixels are the elements of an image that contain information about intensity and color.Pixels are arranged in the form of a matrix. This is known as an RGB image

```python
#importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
pic = plt.imread("image1.jpeg")
```

```python
plt.imshow(pic)
```

```python
#R(0) G(1) B(2)
#filter out image by red channel
v1 = pic[:,:,0]
plt.imshow(v1)
```

```python
#filtering out by the green  channel
#R(0) G(1) B(2)
v2 = pic[:,:,1]
plt.imshow(v2)
```

```python
#filtering out by the blue  channel

v3 = pic[:,:,2]
plt.imshow(v3)
```

```python
#loading in image using pillow library
pic1 = Image.open("pic.jpeg")
pic1 = pic1.resize((400,400))
pic1
```

```python
pic1 = np.array(pic1)
red = pic1.copy()
blue = pic1.copy()
green = pic1.copy()
```

```python
#applying a red filter to our image
red[:,:,(1,2)] = 0
plt.imshow(red)
```

```python
#appying a blue filter to our image
blue[:,:,(0,1)] = 0
plt.imshow(blue)
```

```python
#applying a green filter to our image
green[:,:,(0,2)] = 0
plt.imshow(green)
```

```python
#applying a gray scale to the image
pic1 = pic1[:,:,2]
plt.imshow(pic1,cmap='gray')
```

```python
pic3 = Image.open("pic.jpeg")
pic3 = pic3.resize((400,400))
pic3
```

```python
pic3.rotate(100)
```

```python
print(pic3.format)

print(pic3.size)

print(pic3.mode)
```

```python
cropped_img = pic3.crop((300, 150, 700, 1000))
print(cropped_img.size)

pic3.show()
```

```python
low_res_img = cropped_img.resize((cropped_img.width // 4, cropped_img.height // 4))
low_res_img.show()
```

```python
pic3 = np.array(pic3)
image_flip = cv2.flip(pic3,1)
plt.imshow(image_flip)
```

```python
#pic3 = np.array(pic3)
image_flip2 = cv2.flip(pic3,-1)
plt.imshow(image_flip2)
```

```python
#creating a canvas
canv1 = np.zeros((1000,1000,3))
plt.imshow(canv1)
```

```python
#creating a circle
cv2.circle(canv1,center=(430,500),radius=300,color=(120,40,60),thickness=8)
plt.imshow(canv1)
```

```python
#creating a canvas
canv2 = np.zeros((1000,1000,3))
plt.imshow(canv2)
```

```python
cv2.rectangle(canv2,pt1=(200,200),pt2=(800,800),color=(150,100,60),thickness=-1)
plt.imshow(canv2)
```

```python
#creating a circle
cv2.circle(canv1,center=(430,500),radius=300,color=(120,40,60),thickness=8)
cv2.circle(canv1,center=(200,350),radius=150,color=(0,140,20),thickness=8)
cv2.circle(canv1,center=(200,630),radius=150,color=(50,0,200),thickness=-1)

plt.imshow(canv1)
```

```python
cav4 = np.zeros((1000,1000))

red = cav4.copy()
blue = cav4.copy()
green = cav4.copy()

red[200:400,200:400] = 225
blue[250:450,250:450] = 225
green[300:500,300:500] = 225

red = Image.fromarray(red).convert("L")
blue = Image.fromarray(blue).convert("L")
green = Image.fromarray(green).convert("L")

pic = Image.merge("RGB",[red,blue,green])
plt.imshow(pic)
```
